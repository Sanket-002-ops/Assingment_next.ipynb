Question3:  Explain the following terms:

*   Difference between np.array, np.asarray and np.asanyarray
*   Difference between Deep copy and Shallow copy



```python
#  Explain the following terms: 1 .  The difference in np.array, np.asarray and np.asanarray. And 2.  The difference between Deep copy and Shallow copy.

# 1. Difference between np.array, np.asarray and np.asanyarray

# np.array:
# Always creates a new array object, even if the input is already an array.
# It copies the data from the input to the new array.

# np.asarray:
# Creates a new array object if the input is not an array, or a list, a tuple.
# If the input is already an array, it returns the original array without copying.
# If the input is a list, or a tuple, it creates a new array object containing a copy of the input data

# np.asanyarray:
# Similar to np.asarray but can accept any array-like object.
# If the input is already an array (of any subclass), it returns the original array without copying.

# Example:
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.asarray(arr1)
arr3 = np.asanyarray(arr1)


print("arr1:", arr1, "id:", id(arr1))  # arr1: [1 2 3] id: memory location 1
print("arr2:", arr2, "id:", id(arr2))  # arr2: [1 2 3] id: memory location 2
print("arr3:", arr3, "id:", id(arr3))  # arr3: [1 2 3] id: memory location 1

# 2. Difference between Deep copy and Shallow copy

# Shallow copy: creates a new object, but the elements of the new object refer to the same memory locations as the original object.
# Modifying the elements of the copied object will affect the original object and vice versa.


# Deep copy: creates a completely independent copy of the original object and its elements.
# Modifying the elements of the copied object will not affect the original object, and vice versa.


import copy


original_list = [[1, 2, 3], [4, 5, 6]]

# Shallow copy
shallow_copy = copy.copy(original_list) #or shallow_copy = original_list[:]
#Deep copy
deep_copy = copy.deepcopy(original_list)

shallow_copy[0][0] = 99
print("Original list:", original_list)
print("Shallow copy:", shallow_copy)
print("Deep copy:", deep_copy)
# Output:
#Original list: [[99, 2, 3], [4, 5, 6]]
#Shallow copy: [[99, 2, 3], [4, 5, 6]]
#Deep copy: [[1, 2, 3], [4, 5, 6]]


deep_copy[0][0] = 100
print("Original list:", original_list) # Original list remains unchanged
print("Deep copy:", deep_copy) # Only deep copy changes

```

```python
# Create a numpy array with random integers between 1 and 10 of shape (5,6). After creating the array perform the following operations.  1: Extract all even integers from array . and 2: Extract all odd integers from array.

import numpy as np

# Create a numpy array with random integers between 1 and 10 of shape (5,6)
random_array = np.random.randint(1, 11, size=(5, 6))

# Extract all even integers from the array
even_numbers = random_array[random_array % 2 == 0]

# Extract all odd integers from the array
odd_numbers = random_array[random_array % 2 != 0]

print("Original array:\n",random_array)
print("\nEven numbers:\n", even_numbers)
print("\nOdd numbers:\n", odd_numbers)
```

Question7: Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type. Also display the table attributes and data types of each column.

```python
# clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type. Also display the table attributes and data types of each column.

import pandas as pd
import numpy as np

# Sample data (replace with your actual data)
data = {'Phone': ['(123) 456-7890', '123-456-7890', '+1 123 456-7890', 'abc123xyz', np.nan]}
df = pd.DataFrame(data)

# Clean and transform the 'Phone' column
def clean_phone(phone):
    if pd.isna(phone):
        return np.nan
    phone = ''.join(filter(str.isdigit, str(phone)))
    if phone:
      return int(phone)
    return np.nan

df['Phone_Cleaned'] = df['Phone'].apply(clean_phone)
print(df)

# Display table attributes and data types
print("\nTable Attributes and Data Types:")
print(df.info())
```

# Question15: Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
*   a):  Plot a histogram of the data with 30 bins.
*   b):  Overlay a line plot representing the normal distribution's probability density function (PDF).
*   c):   Label the x-axis as Value and the y-axis as Frequency/Probability.
*   d): Set the title of the plot as Histogram with PDF Overlay.



```python
# Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create the data
data = np.random.normal(loc=0, scale=1, size=1000)

# a) Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', label='Histogram')

# b) Overlay the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

# c) Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title
plt.title('Histogram with PDF Overlay')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

# Question17: Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin(quadrants), add a legend , label the axes, and set the title as Quadrant-wise Scatter Plot.

```python
# Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin(quadrants), add a legend , label the axes, and set the title as Quadrant-wise Scatter Plot.

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Create two random arrays
x = np.random.rand(100)
y = np.random.rand(100)

# Determine quadrant for each point
quadrants = []
for i in range(len(x)):
    if x[i] >= 0.5 and y[i] >= 0.5:
        quadrants.append('Quadrant I')
    elif x[i] < 0.5 and y[i] >= 0.5:
        quadrants.append('Quadrant II')
    elif x[i] < 0.5 and y[i] < 0.5:
        quadrants.append('Quadrant III')
    else:
        quadrants.append('Quadrant IV')

# Create the scatter plot
sns.scatterplot(x=x, y=y, hue=quadrants)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Quadrant-wise Scatter Plot")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

#Question18: With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as Sine Wave Function.

```python
#  With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as Sine Wave Function.

import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Output the plot in the notebook
output_notebook()

# Create a figure object
p = figure(title="Sine Wave Function", x_axis_label="x", y_axis_label="sin(x)")

# Generate x values from 0 to 2*pi
x = np.linspace(0, 2*np.pi, 100)

# Calculate sine values
y = np.sin(x)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Add grid lines
p.grid.grid_line_alpha=0.5

# Show the plot
show(p)
```

# Question19: Using Bokeh , generate a bar chart of randomly generated categorical data , color bars based on their values , add hover tooltips to display exact values , label the axes , and set the title as Random Categorical Bar Chart.

```python
# Using Bokeh , generate a bar chart of randomly generated categorical data , color bars based on their values , add hover tooltips to display exact values , label the axes , and set the title as Random Categorical Bar Chart.

from bokeh.plotting import figure, show, output_notebook
import pandas as pd
import numpy as np
from bokeh.models import LinearColorMapper

# Output the plot in the notebook
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 101, size=len(categories))

# Create a Pandas DataFrame for the data
data = {'Category': categories, 'Value': values}
df = pd.DataFrame(data)

# Create a Bokeh figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add a bar renderer with color based on values
color_mapper = bokeh.models.LinearColorMapper(palette="Viridis256", low=min(values), high=max(values))
p.vbar(x='Category', top='Value', width=0.9, source=bokeh.models.ColumnDataSource(df),
       line_color='white', fill_color={'field': 'Value', 'transform': color_mapper})


# Add hover tooltips
hover = bokeh.models.HoverTool(tooltips=[("Category", "@Category"), ("Value", "@Value")])
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = "Category"
p.yaxis.axis_label = "Value"

# Show the plot
plt.show(p)
```
